```python
import requests
import numpy as np
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://namu.wiki/w/%EC%9D%BC%EB%B3%B8%EC%9D%98%20%EC%83%81%EC%9A%A9%ED%95%9C%EC%9E%90/%EB%AA%A9%EB%A1%9D"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
table = soup.find('table')
print(table)
```

```python
kanji_data = pd.DataFrame(columns=["Num", "Kanji", "Old", "Kor", "Jap", "Ex"])
kanji_list = []
```

```python
kanji_data = pd.DataFrame(columns=["Num", "Kanji", "Old", "Kor", "Jap", "Ex"])
kanji_list = []
for row in table.find("tbody").find_all('tr'):
    col = row.find_all("td")
    kanjiNum = col[0].text
    kanji = col[1].text
    kanjiOld = col[2].text
    audioKor = col[3].text
    audioJap = col[4].text
    audioJap_list = []
    for string in col[4].strings:
        audioJap_list.append(string)
    ex_list = []
    for string in col[5].strings:
        ex_string = string.split(sep='、')
        ex_list.append(ex_string)
    ex = col[5].text
    kanji_list.append([kanjiNum, kanji, kanjiOld, audioKor, audioJap_list, ex_list])
kanji_data = pd.DataFrame(kanji_list, columns=["Num", "Kanji", "Old", "Kor", "Jap", "Ex"])
kanji_data
```
